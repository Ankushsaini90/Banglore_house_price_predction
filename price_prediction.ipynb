# square fit

```python
df.head()
```

```python
df["total_sqft"].unique()
```

we have some values in range like "1133 - 1384". So we have to covert this into single value.

```python
def convertRange(x):
  temp = x.split("-")
  if len(temp) == 2:
    return (float(temp[0]) + float(temp[1]))/2
  try:
    return float(x)
  except:
    return
```

```python
df["total_sqft"]=df["total_sqft"].apply(convertRange)
```

```python
df["total_sqft"].unique()
```

# price per sqft

```python
df["price_per_sqft"] = df["price"]*100000 / df["total_sqft"]
```

# outlier

```python
# location
df["location"].value_counts()
```

we have some location count 1. so we put them together as "other" location.

we have 1305 locations

```python
df["location"] = df["location"].apply(lambda x:x.strip())
location_count = df["location"].value_counts()
```

```python
location_count
```

now after using Strip(remove space) we have 1294 rows.

```python
location_less_10 = location_count[location_count <= 10]
location_less_10
```

```python
# location which are less than 10, will consider as other
df["location"] = df["location"].apply(lambda x: "other" if x in location_less_10 else x)
```

```python
df["location"].value_counts()
```

```python
df.describe()
```

```python
(df["total_sqft"]/df["bhk"]).describe()
```

square_fit per bhk goes minimun to 0.25 which is not making sense so we consedir them as outlier.

```python
df = df[((df["total_sqft"]/df["bhk"])>=300)]
df.describe()
```

```python
# cosdering pricr_per_sqft as (mean+std) or (mean-std)
def sqft_outlier_remover(df):
  df_output = pd.DataFrame()
  for key, subdf in df.groupby("location"):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
    df_output = pd.concat([df_output,gen_df],ignore_index = True)
  return df_output
df = sqft_outlier_remover(df)
df.describe()
```

```python

def bhk_outlier_remover(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby("location"):
    bhk_stats = {
        "mean": np.mean(location_df.bhk),
        "std": np.std(location_df.bhk),
        "count": location_df.bhk.count()
    }

    for bhk, bhk_df in location_df.groupby("bhk"):
      stats = bhk_stats.get(bhk-1)
      if stats and stats["count"] > 5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats["mean"])].index.values)
  return df.drop(exclude_indices, axis = "index")

```

```python
df = bhk_outlier_remover(df)
df
```

```python
X = df.drop(columns=["price"])
y = df["price"]
```

# model

```python
# importing model libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
# spliting
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## linear regression

```python
col_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ["location"]),
        remainder="passthrough"
        )
```

```python
scaler = StandardScaler()
```

```python
pipeline = make_pipeline(col_trans, scaler, LinearRegression())
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred_lr = pipeline.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

## lasso

```python
pipe = make_pipeline(col_trans, scaler, Lasso())
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lasso )
```

## ridge

```python
pipe = make_pipeline(col_trans, scaler, Ridge())
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipeline.predict(X_test)
```

```python
r2_score(y_test, y_pred_ridge)
```

# result


```python
print("Linear Regression: ", r2_score(y_test, y_pred_lr))
print("Linear Regression(Lasso): ", r2_score(y_test, y_pred_lasso))
print("Linear Regression(ridge): ", r2_score(y_test, y_pred_ridge))
```
